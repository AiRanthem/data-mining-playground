```python
import pandas as pd
import numpy as np
import timeit
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
df.head()
```

```python
df.info()
```

```python
allTransection = []
datalen = len(df)
totalCol = len(df.columns)
```

```python
totalCol
```

```python
# making basket
# -------------
```

```python
for i in range(0, datalen):
    allTransection.append([str(df.values[i, j]) for j in range(0, totalCol)])
```

```python
allTransection
```

```python
import pyfpgrowth as pfp
```

```python
start = timeit.timeit()
```

```python
pattern = pfp.find_frequent_patterns( allTransection, 5 )
rules   = pfp.generate_association_rules( pattern, 0.8 ) 

end = timeit.timeit()
```

```python
len(pattern)
```

```python
print('times for finding patterns & rules: ', end - start)
```
