```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = pd.Series([1,3,5,6,np.nan,8])  # Series: 1D array 
s
```

```python
#date_range在处理时间序列数据时产生一个DatetimeIndex，就是时间序列数据的索引。
#pandas.date_range(start=None, end=None, periods=None, freq=’D’, tz=None, normalize=False, name=None, closed=None, **kwargs)

dates = pd.date_range('20200301', periods=6)  #产生6个时间序列数据的索引
dates
```

```python
#dataframe是一个类似表的结构，由多个Series组成，而Series在dataframe中叫做columns
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list('ABCD')) 
df
```

```python
df.index
```

```python
df.columns
```

```python
df.describe()
```

```python
df.T
```

```python
df.sort_index(ascending=False) #默认axis=0,0按照行名排序；1按照列名排序;默认 ascending=True
```

```python
df.sort_index(axis=1, ascending=False)
```

```python
df.sort_values(by='B') #值排序   按列B的值从小排列到大,默认是升序，否则 ascending=False
```

```python
df.sort_values(by=['A','C'])
```

```python
print(df['A'])  #选择单列，结果产生一个Series，等价于df.A
```

```python
df[0:3] #通过[]，对行进行切片，选择前3行.注意，此处不包括右值
```

```python
df['20200302':'20200304'] #通过索引进行切片,注意，此处包括右值
```

```python
df2 = df.copy()# 使用isin()方法来过滤数据
df2['E'] = ['one','one','two','three','four','three']
df2
```

```python
# 通过dict对象来创建像series一样的Dataframe。Dataframe的行为索引index，列为labels
df3 = pd.DataFrame({ 'A' : 1.,
                    'B' : pd.Timestamp('20130102'),
                    'C' : pd.Series(1,index=list(range(4)),dtype='float32'),
                    'D' : np.array([3] * 4,dtype='int32'),
                    'E' : pd.Categorical(["test","train","test","train"]),
                    'F' : 'foo' })
df3
```

```python
df3.to_excel('df3.xlsx')
```

```python
df4 = pd.DataFrame({"id":[1,2,3,4,5,6], "raw_grade":['a', 'b', 'b', 'a', 'a', 'e']})
df4
```

```python
df4['grade'] = df4['raw_grade'].astype('category') #将数据转换成分类数据
df4
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2020', periods=1000))
print(ts)
```

```python
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index,
   .....:                   columns=['A', 'B', 'C', 'D'])
df = df.cumsum()
plt.figure(); df.plot();  plt.legend(loc='best')
```
